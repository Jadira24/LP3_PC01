IMPORTAR LIBRERÍAS

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

CARGAR EL DATASET

```python
# Cargar el archivo .csv separado por tabulaciones
dataframe = pd.read_csv('HEPATITISCVirusPacientesDeEgiptoDatosPerdidos.csv', sep= ';')
dataframe.head(10)
```

```python
DEFINIR VARIABLES
```

```python
#Variable dependiente 
#datos binarios=(1; no tiene fiebre,  2;tiene fiebre)
y = dataframe['FEVER']

#Variable independiente
x = dataframe.drop(columns=['FEVER'])
```

VISUALIZAR LA DISTRIBUCION DE LAS VARIABLES

```python
sb.catplot(x='FEVER', data=dataframe, kind='count', height=4, aspect=2)
plt.xticks(rotation=0)
plt.show()
```

SEPARAR EN DATOS DE ENTRENAMIENTO Y PRUEBA

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=30)
```

ENTRENAR MODELO KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train)
```
